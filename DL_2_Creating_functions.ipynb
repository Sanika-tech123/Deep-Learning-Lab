```python
# Import libraries
import tensorflow as tf
from tensorflow.keras.datasets import mnist
import matplotlib.pyplot as plt

# Load MNIST dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Show one image from dataset
plt.imshow(X_train[0], cmap='gray')
plt.title("Sample MNIST Image")
plt.show()

# ---------------------------------------------------
# 1. Mean Squared Error (MSE)
# ---------------------------------------------------

# Actual values
y_true_mse = [10, 20, 30]

# Predicted values
y_pred_mse = [12, 18, 33]

# Create MSE loss function
mse = tf.keras.losses.MeanSquaredError()

# Calculate loss
mse_loss = mse(y_true_mse, y_pred_mse)

print("MSE Loss:", mse_loss.numpy())

# ---------------------------------------------------
# 2. Binary Crossentropy
# ---------------------------------------------------

# Actual binary values
y_true_bce = [1, 0, 1]

# Predicted values
y_pred_bce = [0.9, 0.2, 0.8]

# Create Binary Crossentropy loss
bce = tf.keras.losses.BinaryCrossentropy()

# Calculate loss
bce_loss = bce(y_true_bce, y_pred_bce)

print("Binary Crossentropy Loss:", bce_loss.numpy())

# ---------------------------------------------------
# 3. Categorical Crossentropy
# ---------------------------------------------------

# Actual multiclass values
y_true_cce = [[1, 0, 0],
              [0, 1, 0],
              [0, 0, 1]]

# Predicted probabilities
y_pred_cce = [[0.8, 0.1, 0.1],
              [0.2, 0.7, 0.1],
              [0.1, 0.2, 0.7]]

# Create categorical crossentropy loss
cce = tf.keras.losses.CategoricalCrossentropy()

# Calculate loss
cce_loss = cce(y_true_cce, y_pred_cce)

print("Categorical Crossentropy Loss:", cce_loss.numpy())

# ---------------------------------------------------
# Graph for Loss Comparison
# ---------------------------------------------------

loss_names = ['MSE', 'Binary CE', 'Categorical CE']
loss_values = [
    mse_loss.numpy(),
    bce_loss.numpy(),
    cce_loss.numpy()
]

# Create bar graph
plt.bar(loss_names, loss_values)

# Graph title
plt.title("Comparison of Loss Functions")

# Axis labels
plt.xlabel("Loss Functions")
plt.ylabel("Loss Value")

# Show graph
plt.show()
```
